# Hexagonal Node Distribution for RBF-Methods

1. $N$ is the total number of nodes
2. lft and rht are the left and right boundary values of $x$
3. top and bot are the top and bottom boundary values of $y$

```julia
using LinearAlgebra

function hex(N,lft,rht,bot,top)
    
    Rad=sqrt(3)/2; # Radius
    
    x1=0:1:round(sqrt(N)-1);
    
    n=size(x1,1);
    
    X=repeat(transpose(x1),n,1);
   
    Y=repeat(x1,1,n);
   
    
    X=Rad*X;
    
    
    X=X/maximum(X);
    
    Y=Y+repeat(transpose([0,0.5]),n,round(Int,n/2.0) );
    
    Y=Y/maximum(Y);
    
   delta_x=rht-lft;
    
   delta_y=top-bot;
 
   
    x=(X[:]*delta_x)+lft*ones(n^2,1);
    y=(Y[:]*delta_y)+bot*ones(n^2,1);
    
    return(x,y)
    
end

```

```julia
x,y=hex(100,0,2,-1,1)
using Plots

scatter(x,y,markersize = 0.5,title="Hexagonal-Node Distribution", label ="Nodes",xlabel="x-axis",ylabel="y-axis")
```
